#**3. Data Pre-processing**

```python
df.duplicated().sum()
```

manually caping the outliers of weekly sales to 300000 for more robust and bettter presict

```python
dfp=df.copy()
```

```python
dfp['Weekly_Sales'] = np.where(df['Weekly_Sales'] > 3000000, 3000000, df['Weekly_Sales'])
```

```python
sns.histplot(dfp['Weekly_Sales'], kde=True)
```

Capping the outlier from temprature column using IQR method

```python
sns.histplot(df['Temperature'], kde=True)
```

```python

# Calculate the IQR for Temperature
Q1 = dfp['Temperature'].quantile(0.25)
Q3 = dfp['Temperature'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Cap the outliers using np.where
dfp['Temperature'] = np.where(dfp['Temperature'] < lower_bound, lower_bound, dfp['Temperature'])
dfp['Temperature'] = np.where(dfp['Temperature'] > upper_bound, upper_bound, dfp['Temperature'])
```

```python
sns.histplot(dfp['Temperature'], kde=True)
```

converting date to date time format

```python
df['Date']
```

```python
dfp['Date'] = pd.to_datetime(dfp['Date'], format='%d-%m-%Y')
```

```python
dfp['Date']
```

Extracting year, months and dates as features from the dataset

```python
dfp['year'] = dfp['Date'].dt.year
dfp['month'] = dfp['Date'].dt.month
dfp['quarter'] = dfp['Date'].dt.quarter
```

```python
dfp['Week_of_Month'] = ((dfp['Date'].dt.day - 1) // 7) + 1
```

```python
df.head(20)
```

```python
# Holiday dates dictionary
holidays = {
    'Super_Bowl': ['2010-02-12', '2011-02-11', '2012-02-10', '2013-02-08'],
    'Labor_Day': ['2010-09-10', '2011-09-09', '2012-09-07', '2013-09-06'],
    'Thanksgiving': ['2010-11-26', '2011-11-25', '2012-11-23', '2013-11-29'],
    'Christmas': ['2010-12-31', '2011-12-30', '2012-12-28', '2013-12-27']
}

# Convert holiday dates to datetime
for holiday, dates in holidays.items():
    holidays[holiday] = pd.to_datetime(dates)

# Add columns for each holiday type with 0 as default
for holiday in holidays:
    dfp[holiday] = dfp['Date'].isin(holidays[holiday]).astype(int)
```

# **6. Predictive Modelling**

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, VotingRegressor
from sklearn.metrics import r2_score
```

```python
models = {
    'Multiple Linear Regression': LinearRegression(),
    'Decision Tree Regressor': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(random_state=42),
    'Ensemble (Voting Regressor)': VotingRegressor(estimators=[
        ('lr', LinearRegression()),
        ('dt', DecisionTreeRegressor(random_state=42)),
        ('rf', RandomForestRegressor(random_state=42))
    ])
}

# Step 4: Train models and evaluate R² scores
r2_scores = {}
for model_name, model in models.items():
    model.fit(x_train_scaled, y_train)  # Train the model
    y_pred = model.predict(x_test_scaled)  # Make predictions
    r2 = r2_score(y_test, y_pred)  # Calculate R² score
    r2_scores[model_name] = r2  # Store the R² score

# Step 5: Display R² scores
for model_name, r2 in r2_scores.items():
    print(f"{model_name}: R² Score = {r2:.4f}")
```

# **Random Forrest gets the higest R2 score for this data type so we will be moving forward with Random Forest**

```python
rf_model = RandomForestRegressor(random_state=42)
```

```python
rf_model.fit(x_train_scaled, y_train)
```

```python
y_predict = rf_model.predict(x_test_scaled)
y_predict
```

```python
rf_model.score(x_train_scaled, y_train)
```

```python
# Model evaluation
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_predict)

r2 = r2_score(y_test, y_predict) #how well our model fits the data

print(f'Mean Squared Error: {mse}')

print(f'R-squared: {r2}')
```
